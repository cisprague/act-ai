```python
import torch, numpy as np, matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
%matplotlib
```

```python
n = 50
net = torch.nn.Sequential(
    torch.nn.Linear(7, n),
    torch.nn.ReLU(),
    *[
    torch.nn.Linear(n, n),
    torch.nn.ReLU()
    ]*5,
    torch.nn.Linear(n, 1),
    torch.nn.Sigmoid()
).double()
```

```python
lf = torch.nn.MSELoss()
```

```python
lr = 1e-4
```

```python
ogdata = np.load("../npy/moc_data_cart.npy")
data   = np.copy(ogdata)
np.random.shuffle(data)
data = data[:200]
ogdata = ogdata[:200]
```

```python
ogx, ogy = ogdata[:, :7], ogdata[:, 7]
x, y = data[:, :7], data[:, 7]
```

```python
scaler = StandardScaler().fit(x)
x = scaler.transform(x)
ogx = scaler.transform(ogx)
```

```python
ogx = torch.autograd.Variable(torch.from_numpy(ogx))
ogy = torch.autograd.Variable(torch.from_numpy(ogy))
x = torch.autograd.Variable(torch.from_numpy(x))
y = torch.autograd.Variable(torch.from_numpy(y))
```

```python
epos = 5000
```

```python
optimiser = torch.optim.Adam(net.parameters(), lr=lr)
```

```python
lds = list()
```

```python
for epo in range(epos):
    
    yp = net(x)
    
    l = lf(yp, y)
    ld = l.data[0]
    lds.append(ld)
    print(epo, l.data[0])
    
    optimiser.zero_grad()
    
    l.backward()
    
    optimiser.step()
```

```python
fig, axs = plt.subplots(2, 1)

axs[0].plot(y.data.numpy(), "k.")
axs[0].plot(net(x).data.numpy(), "x")
axs[1].plot(lds, "k-")
plt.show()
```
