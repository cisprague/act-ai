```python
import sys
sys.path.append("../py")
from trajectory import p2p
import pygmo as pg
import pygmo_plugins_nonfree as pg7
```

```python
udp = p2p(nseg=50)
prob = pg.problem(udp)
print(prob)
```

```python
uda = pg7.snopt7(True, "/usr/lib/libsnopt7_c.so")
algo = pg.algorithm(uda)
uda = pg.mbh(algo, 5)
algo = pg.algorithm(uda)
```

```python
pop = pg.population(prob, 1)
```

```python
pop = algo.evolve(pop)
```

```python
udp.plot(pop.champion_x)
```
